```python

import pandas as pd

# Load dataset
df = pd.read_csv("dirty_student_data.csv")

print("Original Data Shape:", df.shape)

# Log dictionary
log = {}

# 1. Missing values
missing_before = df.isnull().sum().sum()
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
missing_after = df.isnull().sum().sum()
log['Missing Values Filled'] = missing_before - missing_after

# 2. Remove duplicates
duplicates_before = df.duplicated().sum()
df = df.drop_duplicates()
duplicates_after = df.duplicated().sum()
log['Duplicates Removed'] = duplicates_before - duplicates_after

# 3. Fix string inconsistencies
df['Name'] = df['Name'].str.strip().str.lower()
df['City'] = df['City'].str.strip().str.lower()

log['String Cleaned'] = "Name & City standardized"

# 4. Fix data types
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Marks'] = pd.to_numeric(df['Marks'], errors='coerce')

log['Data Types Fixed'] = "Date & Marks converted"

# 5. Handle outliers (capping)
Q1 = df['Marks'].quantile(0.25)
Q3 = df['Marks'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

before_outliers = ((df['Marks'] < lower_bound) | (df['Marks'] > upper_bound)).sum()

df['Marks'] = df['Marks'].clip(lower_bound, upper_bound)

after_outliers = ((df['Marks'] < lower_bound) | (df['Marks'] > upper_bound)).sum()

log['Outliers Treated'] = before_outliers - after_outliers

# Final validation
print("\nFinal Data Shape:", df.shape)
print("\nCleaning Log:")
for k, v in log.items():
    print(f"{k}: {v}")

# Save clean file
df.to_csv("clean_student_data.csv", index=False)

print("\nCleaned dataset saved successfully!")
```
