```python
# ============================================
# 📌 TITANIC DATA ANALYSIS PROJECT
# Matplotlib + Seaborn 
# ============================================

# 🔷 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Style set (professional look)
sns.set_theme(style="darkgrid")

# 🔷 2. Load Dataset
df = pd.read_csv("titanic_professional.csv")

# 🔷 3. Basic Data Overview
print("\n🔹 First 5 Rows:")
print(df.head())

print("\n🔹 Dataset Info:")
print(df.info())

print("\n🔹 Statistical Summary:")
print(df.describe())

print("\n🔹 Missing Values:")
print(df.isnull().sum())


# ============================================
# 🔥 DATA VISUALIZATION (Seaborn + Matplotlib)
# ============================================

# 🔷 4. Survival Count
plt.figure()
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()


# 🔷 5. Survival by Gender
plt.figure()
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.show()


# 🔷 6. Age Distribution
plt.figure()
sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()


# 🔷 7. Passenger Class vs Survival
plt.figure()
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Class vs Survival")
plt.show()


# 🔷 8. Fare Distribution (Matplotlib Required)
plt.figure()
plt.hist(df["Fare"])
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()


# 🔷 9. Fare vs Survival (Outliers Detection)
plt.figure()
sns.boxplot(x="Survived", y="Fare", data=df)
plt.title("Fare vs Survival")
plt.show()


# 🔷 10. Correlation Heatmap (ML Insight)
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.show()


# 🔷 11. Pairplot (Advanced Visualization)
sns.pairplot(df, hue="Survived")


# 🔷 12. Embarked vs Survival
plt.figure()
sns.countplot(x="Embarked", hue="Survived", data=df)
plt.title("Port of Embarkation vs Survival")
plt.show()


# 🔷 13. Age Groups Analysis (Feature Engineering)
df["Age_Group"] = pd.cut(df["Age"], bins=[0, 12, 19, 35, 60, 80],
                         labels=["Child", "Teen", "Young Adult", "Adult", "Senior"])

plt.figure()
sns.countplot(x="Age_Group", hue="Survived", data=df)
plt.title("Age Group vs Survival")
plt.xticks(rotation=30)
plt.show()


# 🔷 14. Average Fare by Class
avg_fare = df.groupby("Pclass")["Fare"].mean()
print("\n🔹 Average Fare by Class:")
print(avg_fare)


# 🔷 15. Survival Rate Calculation
survival_rate = df["Survived"].mean() * 100
print(f"\n🔹 Overall Survival Rate: {round(survival_rate, 2)}%")


```
